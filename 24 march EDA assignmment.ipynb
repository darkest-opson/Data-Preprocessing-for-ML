Q2. How did you handle missing data in the wine quality data set during the feature engineering process?
Discuss the advantages and disadvantages of different imputation techniques.

In wine quality dataset there was no missing values.

Here are some of the most common imputation techniques and their pros and cons:

Mean or median imputation:

This technique replaces missing values with the mean or median of the non-missing values in the variable.

Advantages:

Easy to implement.

Preserves the sample size.

Disadvantages:

May lead to biased estimates if the missing values are not missing at random (MAR).

Does not account for any relationships between variables.

Regression imputation:

This technique uses a regression model to predict the missing values based on the other variables in the dataset.

Advantages:

Accounts for relationships between variables.

Can produce more accurate estimates than mean or median imputation.

Disadvantages:

Assumes a linear relationship between variables.

Requires a significant amount of data and computation.

K-nearest neighbors (KNN) imputation:

This technique fills in missing values with the average value of the K nearest data points.

Advantages:

Accounts for relationships between variables.

Can handle non-linear relationships between variables.

Disadvantages:

Can be computationally expensive.

The value of K can significantly affect the results.

Multiple imputation:

This technique generates multiple imputed datasets by filling in missing values with different estimated values 
based on the distribution of the non-missing values.

Advantages:

Accounts for uncertainty in the imputation process.

Produces more accurate estimates than single imputation methods.

Disadvantages:

Can be computationally expensive.

Requires a significant amount of data.

Maximum likelihood estimation:

This technique estimates the missing values by maximizing the likelihood of the observed data.

Advantages:

Accounts for relationships between variables.

Produces unbiased estimates under the assumption of missing completely at random (MCAR) or MAR.

Disadvantages:

Assumes a specific probability distribution for the data.

Can be computationally expensive.

Q5. Load the wine quality data set and perform exploratory data analysis (EDA) to identify the distribution
of each feature. Which feature(s) exhibit non-normality, and what transformations could be applied to
these features to improve normality?

To perform EDA on the wine quality dataset, we first need to load the dataset and import the necessary libraries.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load dataset
wine = pd.read_csv('winequality-red.csv')

# View first few rows of the dataset
wine.head()
```

```python
# Now, let's visualize the distribution of each feature using histograms and density plots

# Create subplots for each feature
fig, ax = plt.subplots(nrows=2, ncols=6, figsize=(16, 6))
ax = ax.flatten()

# Plot histograms and density plots for each feature
for i, col in enumerate(wine.columns[:-1]):
    sns.histplot(wine[col], kde=True, ax=ax[i], color='b')
    sns.kdeplot(wine[col], ax=ax[i], color='r')

# Remove empty subplots
for i in range(11, len(ax)):
    fig.delaxes(ax[i])

plt.tight_layout()
plt.show()
```

From the plots, we can see that several features exhibit non-normality, including:

fixed acidity

volatile acidity

citric acid

residual sugar

chlorides

free sulfur dioxide

total sulfur dioxide

density

pH

To improve normality, we can apply different transformations, depending on the type of non-normality. Here are some examples:

Log transformation - for right-skewed distributions, such as fixed acidity, volatile acidity, citric acid, and residual sugar.

```python
# Log transform features
wine['fixed acidity'] = np.log(wine['fixed acidity'])
wine['volatile acidity'] = np.log(wine['volatile acidity'])
wine['citric acid'] = np.log(wine['citric acid'])
wine['residual sugar'] = np.log(wine['residual sugar'])

# Replot distributions
fig, ax = plt.subplots(nrows=2, ncols=6, figsize=(16, 6))
ax = ax.flatten()

for i, col in enumerate(wine.columns[:-1]):
    sns.histplot(wine[col], kde=True, ax=ax[i], color='b')
    sns.kdeplot(wine[col], ax=ax[i], color='r')

for i in range(11, len(ax)):
    fig.delaxes(ax[i])

plt.tight_layout()
plt.show()
```

```python
# Square root transformation - for left-skewed distributions, such as free sulfur dioxide, total sulfur dioxide, and pH.

# Square root transform features
wine['free sulfur dioxide'] = np.sqrt(wine['free sulfur dioxide'])
wine['total sulfur dioxide'] = np.sqrt(wine['total sulfur dioxide'])
wine['pH'] = np.sqrt(wine['pH'])

# Replot distributions
fig, ax = plt.subplots(nrows=2, ncols=6, figsize=(16, 6))
ax = ax.flatten()

for i, col in enumerate(wine.columns[:-1]):
    sns.histplot(wine[col], kde=True, ax=ax[i], color='b')
    sns.kdeplot(wine[col], ax=ax[i], color='r')

for i in range(11, len(ax)):
    fig.delaxes(ax[i])

plt.tight_layout()
plt.show()
```

Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of
features. What is the minimum number of principal components required to explain 90% of the variance in
the data?

To perform PCA on the wine quality dataset, we first need to load the dataset and import the necessary libraries.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load dataset
wine = pd.read_csv('winequality-red.csv')

# Separate features and target variable
X = wine.iloc[:, :-1]
y = wine.iloc[:, -1]

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Instantiate PCA with 11 components (number of features)
pca = PCA(n_components=11)

# Fit PCA to standardized features
pca.fit(X_scaled)

# Compute explained variance ratio for each component
explained_var = pca.explained_variance_ratio_

# Plot cumulative explained variance ratio
cumulative_var = np.cumsum(explained_var)
plt.plot(range(1, 12), cumulative_var, marker='o')
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance ratio')
plt.show()
```

From the plot, we can see that the first 7 principal components explain more than 90% of the variance in the data. Therefore, we can use these 7 principal components to reduce the number of features.

```python
# Instantiate PCA with 7 components
pca = PCA(n_components=7)

# Fit PCA to standardized features
X_pca = pca.fit_transform(X_scaled)

# Convert to dataframe
X_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7'])

# Concatenate with target variable
wine_pca = pd.concat([X_pca, y], axis=1)

# View first few rows of the new dataset
wine_pca.head()
```
