```python
#convert dates in numerical form by converting into date month and year column
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df[['Date']].dtypes
```

```python
#date monnth year are in string but it should be numerical form
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Date'].astype(int)
df['Year']=df['Date'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
## Change the Arrival Time into hours and minutes

df['Arrival_Hours'] = df['Arrival_Time'].str[0:5].str.split(':').str[0]
```

```python
df['Arrival_minutes'] = df['Arrival_Time'].str[0:5].str.split(':').str[1]
```

```python
# Type casting into integer
```

```python
df['Arrival_Hours'] = df['Arrival_Hours'].astype(int)
```

```python
df['Arrival_minutes'] = df['Arrival_minutes'].astype(int)
```

```python
df.info()
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Dep_Hours'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_Minutes'] = df['Dep_Time'].str.split(':').str[1]
```

```python
#change into integer type
```

```python
df['Dep_Hours']=df['Dep_Hours'].astype(int)
df['Dep_Minutes']=df['Dep_Minutes'].astype(int)
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.info()
```

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df['Duration_Hours'] = df['Duration'].str[0:].str.split().str[0].str.split('h').str[0]
```

```python
df['Duration_Minutes'] = df['Duration'].str[0:].str.split().str[1].str.split('m').str[0]
```

```python
df[df['Duration'].str[0:] == '5m']
```

```python
df.drop('Duration',axis=1,inplace=True)
```

```python
OHE = OneHotEncoder()
```

```python
OHE_array = OHE.fit_transform(df[['Airline','Source','Destination']]).toarray()
```
